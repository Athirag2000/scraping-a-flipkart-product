```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
```

```python
link='https://dl.flipkart.com/s/PZAWY0uuuN'
```

```python
page=requests.get(link)
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find('span',class_="B_NuCI")
print(name)
```

```python
# to get just the name we will use the below code
name.text
```

```python
#get rating of a product
rating=soup.find('div',class_="_3LWZlK")
print(rating)
rating.text
```

```python
#get other details and specifications of the product
specification=soup.find('div',class_="_2cM9lP")
print(specification)
specification.text
```

```python
#get price of the product
price=soup.find('div',class_='_30jeq3 _16Jk6d')
print(price)
price.text
```
